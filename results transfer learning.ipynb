```python
df=[]
for i in np.arange(24):
    df.append(pd.DataFrame(pd.read_csv('Results/trans_learn_{:03d}.csv'.format(i+1))))
```

```python
for i in np.arange(24):
    print(df[i].base_model[0],df[i].input_image[0])
```

```python
accs=[]
val_accs=[]
loss=[]
val_loss=[]
for_label=[]

for i in np.arange(8):
    accs.append(df[i].acc.values)
    val_accs.append(df[i].val_acc.values)
    loss.append(df[i].loss.values)
    val_loss.append(df[i].val_loss.values)
    for_label.append([df[i].base_model[0],'{:0.5f}'.format(df[i].initial_lr[0])])
    for_input_image=df[i].input_image[0]
```

```python
import matplotlib.colors as colors
import matplotlib.cm as cmx

values = range(8)

fig = plt.figure()
ax = fig.add_subplot(111)

my_cmap = cm = plt.get_cmap('seismic') 
cNorm  = colors.Normalize(vmin=0, vmax=values[-1])
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=my_cmap)
print(scalarMap.get_clim())

lines = []
for idx in range(len(accs)):
    line = accs[idx]
    colorVal = scalarMap.to_rgba(values[idx])
    retLine, = ax.plot(line,
                       color=colorVal,
                       label=for_label[idx])
    lines.append(retLine)
#added this to get the legend to work
handles,labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, loc='lower right', ncol=2)
ax.grid()
plt.ylim([0.3, 1])
plt.title(for_input_image)
plt.show()
```

```python
def print_tl_results(df,metric):
    accs=[]
    val_accs=[]
    loss=[]
    val_loss=[]
    for_label=[]

    for i in np.arange(16,24):
        accs.append(df[i].acc.values)
        val_accs.append(df[i].val_acc.values)
        loss.append(df[i].loss.values)
        val_loss.append(df[i].val_loss.values)
        for_label.append([df[i].base_model[0],'{:0.5f}'.format(df[i].initial_lr[0])])
        for_input_image=df[i].input_image[0]
        
    #========================
    
    x_values = range(len(accs))
    y_values=[accs if metric=='acc' else val_acc if metric=='val_acc' else loss if metric=='loss' else 'val_loss']
    fig = plt.figure()
    ax = fig.add_subplot(111)

    my_cmap = cm = plt.get_cmap('seismic') 
    cNorm  = colors.Normalize(vmin=0, vmax=x_values[-1])
    scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=my_cmap)
    print(scalarMap.get_clim())

    lines = []
    for idx in range(len(x_values)):
        line = y_values[idx]
        colorVal = scalarMap.to_rgba(x_values[idx])
        retLine, = ax.plot(line,
                           color=colorVal,
                           label=for_label[idx])
        lines.append(retLine)
    #added this to get the legend to work
    handles,labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels, loc='lower right', ncol=2)
    ax.grid()
    plt.ylim([0.3, 1])
    plt.title(for_input_image)
    plt.show()
```
